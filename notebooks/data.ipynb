# Loading the Data


## MovieData

To make this easier I created a class that quickly and easily loads the data. It can be found in 

    $ data-mining-project/data_load/movie_data.py
    
You can import it like above, by adding it to your path, and then importing the class.

Next, you can call it and pass it some basic parameters. Mainly the min_genre_frequency cut, which specifies the smallest fraction of occurance a genre has for you to keep it. To keep all genres put this fraction to 0.0.

The class also preps the one hot encoding of the data. It can be accessed by obj.one_hot_genres.

```python
movie_data = MovieData(min_genre_frequency=0.0)
movie_data.latex_print_genres()
movie_data.plot_genres()
```

```python
md = MovieData(min_genre_frequency=0.05)
md.latex_print_genres()
# md.plot_genres(savefig='../final_report/genres.pdf')
md.plot_genres()
```

Check if any movies do not have a genre associated with them.

```python
for i in range(movie_data.one_hot_genres.shape[0]):
    if np.all(movie_data.one_hot_genres[i] == np.zeros(movie_data.one_hot_genres.shape[1])):
        print(i)
```
